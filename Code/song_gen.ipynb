```python
#Ex 2
# Compuneți un cântec simplu clasic (ex. Frère Jacques) într-un singur semnal.
# Vom compune Yankee Doodle Day bc 'Murica.

from matplotlib import pyplot as plt
import numpy as np
import scipy as sci
from scipy.io import wavfile as sciowav
from scipy import signal as sig

def solC3(t):
    return np.sin(4*196*np.pi*t)
def laC3(t):
    return np.sin(4*220.61*np.pi*t)
def siC3(t):
    return np.sin(4*246.94*np.pi*t)    
def doC4(t):   
    return np.sin(4*261.63*np.pi*t)
def re(t):   
    return np.sin(4*293.66*np.pi*t)
def mi(t):   
    return np.sin(4*329.63*np.pi*t)
def fa(t):   
    return np.sin(4*349.23*np.pi*t)
def sol(t):   
    return np.sin(4*392*np.pi*t)
def la(t):
    return np.sin(4*440*np.pi*t)
def si(t):   
    return np.sin(4*493.88*np.pi*t)
def doC5(t):   
    return np.sin(2*523.25*np.pi*t)
def pauza(t):
    return 0*t

frecventaA = 50000

patrime = np.arange(start = 0, stop = 0.25, step = 0.00002)
doime = np.arange(start = 0, stop = 0.5, step = 0.00002)

melodie = np.concatenate([(doC4(patrime/4)), (doC4(patrime/4)), (re(patrime/4)), (mi(patrime/4)), (doC4(patrime/4)), (mi(patrime/4)), (re(doime/4)),
                         (doC4(patrime/4)), (doC4(patrime/4)), (re(patrime/4)), (mi(patrime/4)), (doC4(doime/4)), (siC3(doime/4)),
                         (doC4(patrime/4)), (doC4(patrime/4)), (re(patrime/4)), (mi(patrime/4)), (fa(patrime/4)), (mi(patrime/4)), (re(patrime/4)), (doC4(patrime/4)),
                         (siC3(patrime/4)), (solC3(patrime/4)), (laC3(patrime/4)), (siC3(patrime/4)), (doC4(doime/4)), (doC4(doime/4))])

sciowav.write('Ex2.wav ', frecventaA, melodie)
```

```python
#Ex 3
# [4p] Citiți o partitură la intrare (folosind LilyPond sau formatul propriu) și produceți semnalul ce conține melodia la ieșire prin compunerea tonurilor asociate notelor automat.

from matplotlib import pyplot as plt
import numpy as np
import scipy as sci
from scipy.io import wavfile as sciowav
from scipy import signal as sig

def solC3(t):
    return np.sin(4*196*np.pi*t)
def laC3(t):
    return np.sin(4*220.61*np.pi*t)
def siC3(t):
    return np.sin(4*246.94*np.pi*t)    
def doC4(t):   
    return np.sin(4*261.63*np.pi*t)
def re(t):   
    return np.sin(4*293.66*np.pi*t)
def mi(t):   
    return np.sin(4*329.63*np.pi*t)
def fa(t):   
    return np.sin(4*349.23*np.pi*t)
def sol(t):   
    return np.sin(4*392*np.pi*t)
def la(t):
    return np.sin(4*440*np.pi*t)
def si(t):   
    return np.sin(4*493.88*np.pi*t)
def doC5(t):   
    return np.sin(2*523.25*np.pi*t)
def pauza(t):
    return 0*t

frecventaA = 50000

patrime = np.arange(start = 0, stop = 0.25, step = 0.00002)
doime = np.arange(start = 0, stop = 0.5, step = 0.00002)

intrare = open("ex3-in.txt", "r+")

file = intrare.readline().strip()

melodie = np.array([])

while (file != "eof"):
    
    if(file == "solC3 1/4"):
        melodie = np.concatenate([melodie, solC3(patrime/4)])
    if(file == "laC3 1/4"):
        melodie = np.concatenate([melodie, laC3(patrime/4)]) 
    if(file == "siC3 1/4"):
        melodie = np.concatenate([melodie, siC3(patrime/4)])
    if(file == "doC4 1/4"):
        melodie = np.concatenate([melodie, doC4(patrime/4)])
    if(file == "doC4 1/2"):
        melodie = np.concatenate([melodie, doC4(doime/4)])
    if(file == "re 1/4"):
        melodie = np.concatenate([melodie, re(patrime/4)])
    if(file == "re 1/2"):
        melodie = np.concatenate([melodie, re(doime/4)])
    if(file == "mi 1/4"):
        melodie = np.concatenate([melodie, mi(patrime/4)])
    if(file == "fa 1/4"):
        melodie = np.concatenate([melodie, fa(patrime/4)])
    if(file == "sol 1/4"):
        melodie = np.concatenate([melodie, sol(patrime/4)])
    if(file == "la 1/4"):
        melodie = np.concatenate([melodie, la(patrime/4)])
    if(file == "si 1/4"):
        melodie = np.concatenate([melodie, si(patrime/4)])
    if(file == "doC5 1/4"):
        melodie = np.concatenate([melodie, doC5(patrime/4)])

    file = intrare.readline().strip()
    
intrare.close()

sciowav.write('ex3_out.wav', frecventaA, melodie)
```
